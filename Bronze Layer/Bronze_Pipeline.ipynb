```python
from pyspark.sql.functions import * 
from pyspark.sql.types import * 
```

```python
df = spark.readStream.format("cloudFiles")\
          .option("cloudFiles.format", "csv")\
          .option("cloudFiles.schemaLocation", "abfss://source@intechaccstorage.dfs.core.windows.net/checkpoint_Address")\
          .load("abfss://source@intechaccstorage.dfs.core.windows.net/Address")   
```

```python
df.writeStream.format("parquet")\
    .outputMode("append")\
    .option("checkpointLocation", "abfss://bronze@intechaccstorage.dfs.core.windows.net/checkpoint_Address")\
    .option("path", "abfss://bronze@intechaccstorage.dfs.core.windows.net/Address")\
    .trigger(once=True)\
    .start()
```

# Parametrize to Auto Read Data

```python
dbutils.widgets.text("file_name", '')
```

```python
file_name = dbutils.widgets.get("file_name")
```

```python
df = spark.readStream.format("cloudFiles")\
          .option("cloudFiles.format", "csv")\
          .option("cloudFiles.schemaLocation", f"abfss://source@intechaccstorage.dfs.core.windows.net/checkpoint_{file_name}")\
          .load(f"abfss://source@intechaccstorage.dfs.core.windows.net/{file_name}")   
```

```python
df.writeStream.format("parquet")\
    .outputMode("append")\
    .option("checkpointLocation", f"abfss://bronze@intechaccstorage.dfs.core.windows.net/checkpoint_{file_name}")\
    .option("path", f"abfss://bronze@intechaccstorage.dfs.core.windows.net/{file_name}")\
    .trigger(once=True)\
    .start()
```
