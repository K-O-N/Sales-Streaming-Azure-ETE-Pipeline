```python
from pyspark.sql.functions import * 
from pyspark.sql.types import * 
from delta.tables import DeltaTable
```

```python
product = spark.read.format("parquet").load("abfss://bronze@intechaccstorage.dfs.core.windows.net/Product")
product.limit(5).display()
```

```python
product = product.withColumnRenamed('ProductID', 'Product_id')\
                    .withColumnRenamed('Name', 'Product_name')\
                    .withColumnRenamed('ProductNumber', 'Product_number')\
                    .withColumnRenamed('StandardCost', 'Standard_cost')\
                    .withColumnRenamed('ListPrice', 'List_price')\
                    .withColumnRenamed('ProductCategoryID', 'Product_category_id')\
                    .withColumnRenamed('ProductModelID', 'Product_model_id')\
                    .withColumnRenamed('SellStartDate', 'Sell_start_date')\
                    .withColumnRenamed('SellEndDate', 'Sell_end_date')\
                    .withColumnRenamed('DiscontinuedDate', 'Discontinued_date')\
                    .withColumnRenamed('ModifiedDate', 'Modified_date')

product = product.drop('ThumbNailPhoto','ThumbnailPhotoFileName','rowguid', '_rescued_data')
```

```python
product = product.withColumn("Modified_date", to_date(col("Modified_date")))\
                 .withColumn("Sell_start_date", to_date(col("Sell_start_date")))\
                 .withColumn("Discontinued_date", to_date(col("Discontinued_date")))\
                 .withColumn("Modified_date", to_date(col("Modified_date")))\
                 .withColumn("Weight", col("Weight").cast("float"))\
                 .withColumn("Standard_cost", col("Standard_cost").cast("float"))\
                 .withColumn("List_price", col("List_price").cast("float"))

product.limit(5).display()
                 
```

```python
%sql
CREATE TABLE IF NOT EXISTS databricksintech.silver.product
(
    Product_id STRING,
    Product_name STRING,
    Product_number STRING,
    Color STRING,
    Standard_cost FLOAT,
    List_price FLOAT,
    Size STRING,
    Weight FLOAT,
    Product_category_id STRING,
    Product_model_id STRING,
    Sell_start_date DATE,
    Sell_end_date DATE,
    Discontinued_date DATE,
    Modified_date DATE
) 
USING DELTA
LOCATION 'abfss://silver@intechaccstorage.dfs.core.windows.net/product';
```

```python
# Reference target Delta table
silver_table = DeltaTable.forName(spark, "databricksintech.silver.product")

# Execute the Merge (Upsert)
silver_table.alias("target").merge(
    source = product.alias("source"),
    condition = (
        "target.Product_id = source.Product_id"
        
    )
).whenMatchedUpdateAll() \
 .whenNotMatchedInsertAll() \
 .execute()

print("Done!")
```
